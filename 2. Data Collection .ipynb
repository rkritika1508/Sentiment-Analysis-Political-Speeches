```python
pdf_files_link = []
dates = []
# This function is going to the individual pages of the sessions and then extracting the links of all the pdf files of those days.
def get_pdf_files():
    count = 0
    for link in df['0']:
        count += 1
        url = 'https://eparlib.nic.in' + link
        page = requests.get(url, stream=True, verify=False)
        soup = BeautifulSoup(page.text, 'html.parser')
        sessions_link = soup.findAll(class_='standard')
        temp = 's'
        for tag in sessions_link:
            if tag.find('a') != None:
                temp = tag.find('a')['href']
        pdf_files_link.append(temp)
        table = soup.find('table', {'class': 'itemDisplayTable'})
        if table:
            sess_date = table.findAll('td')[5].text
        else:
            sess_date = 0
        dates.append(sess_date)
        print(count, url, sess_date)

get_pdf_files()
```

```python
d = {'Link' : pdf_files_link, 'Date' : dates}
df = pd.DataFrame(d)
```

```python
df.to_csv('pdf_links.csv')
```
